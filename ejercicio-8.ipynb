# Histograma

```python
import matplotlib.pyplot as plt
print("Setup terminado")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

marcas_clase = ["0.9-1.95", "1.69-3.01", "3.02-4.07", "4.08-5.13", "5.14-6.19", "6.20-7.25"]
frecuencia = [4, 8, 7, 6, 2, 1]
marcas_textos = ["1", "2", "3","4", "5", "6"]

plt.bar(marcas_clase, frecuencia,
        width=0.8,
        edgecolor="k",
        color=["#740505", "#740560", "#B227C6", "#9363D3", "#6372D3", "#63B1D3"])
plt.xticks(marcas_clase, marcas_textos, fontsize= 15,)
plt.title("HISTOGRAMA", fontsize=40)
plt.xlabel("Marcas de clase", fontsize=25)
plt.ylabel("Horas dedicadas al internet", fontsize=25)
plt.grid(True)
plt.show()
```

# Ojiva

```python
plt.figure(figsize = (12, 6)) # Ancho, largo del grafico
marcas_clase = [1, 2, 3, 4, 5, 6]#Datos del eje X
# frecuencia =[0.5, 0.15, 0.20, 0.20, 0.10, 0.30,] #Datos del eje Y
frecuencia =[13.33, 39.99, 63.32, 83.22, 96.65, 100] #Datos del eje Y

# Ajuste en las listas para generar el poligino de frecuencia
datos_x = [0] + marcas_clase 
datos_y = [0] + frecuencia 

#Leyendas para las marcas de clase
marcas_textos = ["1", "2", "3","4", "5", "6"]


# Poligono de frecuencia con:
# - Estilo de line puntedo ( linewidth = 5, color = "r", linestyle = ":"). https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html
# - Marcador con forma triangular (marker = "v"). https://matplotlib.org/stable/api/markers_api.html
plt.plot(datos_x, datos_y,
        linewidth = 5, color = "b", linestyle = ":",
        marker = "*", markersize = 25, markeredgecolor = "black", markerfacecolor = "y")

plt.xticks(marcas_clase , marcas_textos, fontsize= 15)# Cambio de las marcas de clese del eje x de número a texto.
plt.title("Ojiva", fontsize = 40) # Titulo
plt.xlabel("Marcas de clase", fontsize = 25) # Etiqueta eje X
plt.ylabel("Horas dedicadas al internet", fontsize = 25) # Etiqueta eje Y
plt.grid(True)#Cuatricula
plt.show()#Se muestra gráfica en pantalla
```

- La frecuencia relativa acumulada indica que aproximadamente el 83.22% de los jóvenes navegan en Internet durante 4 horas o menos al día.
